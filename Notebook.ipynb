```python
import sys, codecs
from os import path

default_words = path.join('./words.txt')
default_stopwords = path.join( './stopwords.txt')
default_verbs = path.join('./verbs.txt')

def words_list(words_file=default_words):
	with codecs.open(words_file, encoding='utf-8') as words_file:
		items = [line.strip() for line in words_file]
		return [item for item in items]

def stopwords_list(stopwords_file=default_stopwords):
	with codecs.open(stopwords_file, encoding='utf8') as stopwords_file:
		return list(set(map(lambda w: w.strip(), stopwords_file)))
```

```python
compile_patterns = lambda patterns: [(re.compile(pattern), repl) for pattern, repl in patterns]

class TurkishNormalizer(Normalizer):
    def __init__(self, remove_extra_spaces=True, persian_style=True, persian_numbers=True, remove_diacritics=True, affix_spacing=True, token_based=False, punctuation_spacing=True):
       super().__init__(self)
       self._punctuation_spacing = punctuation_spacing
       self._affix_spacing = affix_spacing
       self._token_based = token_based
       if self._token_based:
            lemmatizer = TurkishLemmatizer()
            self.words = lemmatizer.words
            self.verbs = lemmatizer.verbs
            self.tokenizer = TurkishWordTokenizer(join_verb_parts=False)
            self.suffixes = {'لر','لرین','لردن','لار','لارین','لاردان','مک','ماخ','ماق','مخ','ام'}
       if self.affix_spacing:
            self.affix_spacing_patterns = compile_patterns([
              (r'([^ ]*) لر', r'\1‌لر'),  # fix لر space-------   در تركي علامت جمع دو حرف 'لر' و 'لار' می باشد که مثلا پس از آمدن حرف ه باید با نیم فاصله قرار گیرند در این جا عبارت منظمی که نوشته شده برای این است که اگر نشانه ی جمع با یک فاصله جدا شده بود به نیم فاصله تبدیل شود. لازم به ذکر است که ممکن است این علامت چسبیده به خود واژه باشد.
              (r'([^ ]*) لر(دن)', r'\1‌لر\2'),  # fix لردن space-------   ممکن است این علامت جمع به همراه حرف اضافه از در فارسی که معادل آن دن در ترکی است استفاده شود.
              (r'([^ ]*) لر(ین)', r'\1‌لر\2'),  # fix لرین space-------    یا با حرف اضافه ی ین که همان برای است.
              (r'([^ ]*) لار', r'\1‌لار'),  # fix لار space-------   در تركي علامت جمع دو حرف 'لر' و 'لار' می باشد که مثلا پس از آمدن حرف ه باید با نیم فاصله قرار گیرند در این جا عبارت منظمی که نوشته شده برای این است که اگر نشانه ی جمع با یک فاصله جدا شده بود به نیم فاصله تبدیل شود. لازم به ذکر است که ممکن است این علامت چسبیده به خود واژه باشد.
              (r'([^ ]*) لار(دان)', r'\1‌لار\2'),  # fix لاردان space-------   ممکن است این علامت جمع به همراه حرف اضافه از در فارسی که معادل آن دان در ترکی است استفاده شود.
              (r'([^ ]*) لار(ین)', r'\1‌لار\2'),  # fix لارین space-------    یا با حرف اضافه ی ین که همان برای است.
              (r'([^ ]ه) ر', r'\1‌ر'),  # fix بنزه ر space-------    در برخی افعال که با مصوت فتحه تمام میشوند(معمولا به صورت نوشتاری *ه ر) آخر فعل به حرف ر ختم می شود که در اینجا نیز نباید جدا نوشته شود.
              (r'([^ ]) (یم|یام|سن|سان|دی|دیر|دور|یوخ|سیز|دیلار|دیلر) ', r'\1‌\2 '),  # fix افزودن نیم فاصله بین افعالی که صرف انتهایی در آن ها جدا نوشته می شود لازم به ذکر است که در ترکی صرف انتهایی فعل بستگی به فعل دارد و ممکن است چسبیده یا جدا باشد
              (r'([^ ]) (میشم|میشام|میسن|میسان|ییپ|ییب|میشوخ|میسیز|ییپلار|ییبلار|ییبلر|ییپلر) ', r'\1‌\2 '),  # fix افزودن نیم فاصله بین افعالی که صرف انتهایی در آن ها جدا نوشته می شود لازم به ذکر است که در ترکی صرف انتهایی فعل بستگی به فعل دارد و ممکن است چسبیده یا جدا باشد
              (r'([^ ]) (م|ن|سی|میز|ز|لری) ', r'\1‌\2 '),  # fix  معادل پسوند مالکیت فارسی برای لغات مثلا ام ات اش
              (r'([^ ]) (يم|ین|ی|یمیز|یز|لاری|ام) ', r'\1‌\2 '),  # fix  معادل پسوند مالکیت فارسی برای لغات مثلا ام ات اش
              (r'([^ ]ه) (گیم|گین|گی|گیمیز|گیز|گلری)', r'\1‌\2'),  # fix   معادل پسوند مالکیت فارسی برای لغات مثلا ام ات اش پس از واژه ی ختم شده به ه 
              (r'([^ ]) (ییرم|ییرسن|ییر|ییروخ|ییسیز|ییلر)', r'\1‌\2'),  # fix صرف برخی افعال در زمان حال
              (r'([^ ]) لیه (ردیم|ردین|ردی|ردوخ|ردیر|ردیلر)', r'\1‌لیه‌\2'),  # fix صرف برخی افعال در زمان ماضی استمراری
              (r'([^ ]ه) ([^ ]{,3})', r'\1‌\2'),  # fix  برخی واژه ها به دلیل تلفظ پایانی حرف ه به مصوت فتحه حرف آخرشان جدا نوشته می شود
              (r'([^ ]) (لیک|لیگ|لوک|لیق|لوق) ', r'\1‌\2 '),  # fix پسوندهای اسم ساز
              (r'([^ ]) (جک|جاخ|جخ) ', r'\1‌\2 '),  # fix پسوندهای فعل ساز
            ])
```
